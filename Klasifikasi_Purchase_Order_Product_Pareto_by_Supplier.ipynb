# KLASIFIKASI ORDER PER SUPPLIER

```python
import pandas as pd
import numpy as np

# Load the Excel file
try:
    df = pd.read_excel('/content/LISTORDER23122024.xlsx')
except FileNotFoundError:
    print("Error: File not found. Please check the file path.")
    exit()

# Create a color mapping
color_map = {
    'PARETO A': '\033[94m',  # Blue
    'PARETO B': '\033[95m',  # Pink
    'PARETO C': '\033[92m',  # Yellow
    'reset': '\033[0m'  # Reset color
}

# Group by 'SUPPLIER 1'
supplier_groups = df.groupby('SUPPLIER 1')

# Print the results with color coding
print("Supplier Product List with Pareto Classification:")
for supplier, group in supplier_groups:
    print(f"Supplier: {supplier}")

    # Sort the group by 'KATEGORI PARETO'
    sorted_group = group.sort_values('KATEGORI PARETO')

    for _, row in sorted_group.iterrows():
        # Select color based on Pareto class
        color = color_map.get(row['KATEGORI PARETO'], '')
        reset = color_map['reset']

        print(f"{color}  - {row['NAMA BARANG']} ({row['QTY']} {row['SATUAN']}){reset}")
    print()  # Extra line for readability
```

```python
import pandas as pd
import os
from docx import Document
from docx.shared import Pt, RGBColor
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT

def save_pareto_output_to_word(input_file, output_file):
    # Definisi warna untuk setiap kategori Pareto
    color_map = {
        'PARETO A': RGBColor(0, 0, 255),    # Biru
        'PARETO B': RGBColor(128, 0, 128),  # Ungu
        'PARETO C': RGBColor(0, 128, 0)     # Hijau
    }

    try:
        # Baca file Excel
        df = pd.read_excel(input_file)
    except FileNotFoundError:
        print(f"Error: File tidak ditemukan. Periksa path file: {input_file}")
        return
    except Exception as e:
        print(f"Terjadi kesalahan saat membaca file: {e}")
        return

    # Pastikan direktori output ada
    os.makedirs(os.path.dirname(output_file), exist_ok=True)

    # Buat dokumen Word baru
    doc = Document()

    # Atur font default
    style = doc.styles['Normal']
    font = style.font
    font.name = 'Cambria'
    font.size = Pt(10)

    # Tambahkan judul
    title = doc.add_heading('Daftar Produk Supplier dengan Klasifikasi Pareto', level=1)
    title.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER

    # Kelompokkan berdasarkan 'SUPPLIER 1'
    supplier_groups = list(df.groupby('SUPPLIER 1'))

    # Iterasi melalui supplier
    for i, (supplier, group) in enumerate(supplier_groups):
        # Tambahkan nama supplier
        supplier_heading = doc.add_heading(f'Supplier: {supplier}', level=2)
        supplier_heading.alignment = WD_PARAGRAPH_ALIGNMENT.LEFT

        # Urutkan berdasarkan 'KATEGORI PARETO'
        sorted_group = group.sort_values('KATEGORI PARETO')

        # Tambahkan produk untuk setiap supplier
        for _, row in sorted_group.iterrows():
            # Tentukan warna berdasarkan kategori Pareto
            pareto_color = color_map.get(row['KATEGORI PARETO'], RGBColor(0, 0, 0))

            # Tambahkan paragraf untuk setiap produk
            p = doc.add_paragraph()

            # Tambahkan nama produk dengan warna sesuai kategori
            run = p.add_run(f"  - {row['NAMA BARANG']} ({row['QTY']} {row['SATUAN']})")
            run.font.color.rgb = pareto_color
            run.font.bold = True

            # Tambahkan kategori Pareto dalam kurung
            run_pareto = p.add_run(f" [{row['KATEGORI PARETO']}]")
            run_pareto.font.italic = True
            run_pareto.font.size = Pt(9)

        # Tambahkan dua paragraf kosong antar supplier (kecuali supplier terakhir)
        if i < len(supplier_groups) - 1:
            doc.add_paragraph()
            doc.add_paragraph()

    # Simpan dokumen
    doc.save(output_file)
    print(f"Output berhasil disimpan di {output_file}")

# Contoh penggunaan

input_file = input('input nama path file: ')
output_file = 'output/list order 20 jan.docx'

save_pareto_output_to_word(input_file, output_file)
```
